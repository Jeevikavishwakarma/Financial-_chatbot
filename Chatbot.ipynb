```python
import pandas as pd

df = pd.read_csv("final_data.csv")

print(df)

#Pre-defining queries  

total_revenue = df['Total Revenue'].sum()
net_income_diff = df['Net Income'].iloc[-1] - df["Net Income"].iloc[-2]
revenue_2023 = df.loc[df["Year"]== 2023, "Total Revenue"].values[0]
net_income_2023 = df.loc[df["Year"]== 2023,"Net Income"].values[0]

#Responses

responses = {
"What is the total revenue?": f"The total revenue is ${total_revenue:,}.",
"How has net income changed over the last year?": f"Net income changed by ${net_income_diff:,} from last year.",
"What is the revenue in 2023?": f"The revenue in 2023 was ${revenue_2023:,}.",
"What is the net income in 2023?": f"The net income in 2023 was ${net_income_2023:,}.",
}

from flask import Flask, request, render_template_string
import threading

# Create the app
app = Flask(__name__)

@app.route("/", methods=["GET", "POST"])
def chatbot():
    user_input = ""
    response = ""
    if request.method == "POST":
        user_input = request.form.get("query")
        response = responses.get(user_input, "Sorry, I don't understand that question.")
    html = """
    <!DOCTYPE html>
    <html>
    <head>
        <title>Financial Chatbot</title>
        <link href="https://cdn.jsdelivr.net/npm/bootstrap@5.3.0/dist/css/bootstrap.min.css" rel="stylesheet">
        <style>
            body {
                background-color: #f7f9fc;
                padding-top: 60px;
            }
            .chatbox {
                max-width: 600px;
                margin: auto;
                padding: 30px;
                background: white;
                border-radius: 15px;
                box-shadow: 0 0 15px rgba(0,0,0,0.1);
            }
            .response-box {
                margin-top: 20px;
                padding: 15px;
                background: #e8f0fe;
                border-left: 4px solid #007bff;
                border-radius: 8px;
                font-size: 1.1rem;
            }
            .btn-ask {
                background-color: #007bff;
                color: white;
            }
            .btn-ask:hover {
                background-color: #0056b3;
            }
        </style>
    </head>
    <body>
        <div class="chatbox">
            <h3 class="text-center mb-4">ðŸ“Š Financial Chatbot</h3>
            <form method="post">
                <div class="mb-3">
                    <label for="query" class="form-label">Ask a question:</label>
                    <input name="query" class="form-control" id="query" placeholder="e.g., What is the total revenue?">
                </div>
                <button type="submit" class="btn btn-ask">Ask</button>
            </form>
            {% if response %}
            <div class="response-box mt-4">
                <strong>Bot:</strong> {{ response }}
            </div>
            {% endif %}
        </div>
    </body>
    </html>
"""
    return render_template_string(html, response=response)

# Function to run Flask in background thread
def run_flask():
    app.run(port=5000)

# Start Flask
thread = threading.Thread(target=run_flask)
thread.start()
```
